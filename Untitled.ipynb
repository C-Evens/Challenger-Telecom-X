## Extraccion de datos

```python
import pandas as pd
import requests
import json
```

```python
url="https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
archivo_datos=requests.get(url)
```

```python
datos=pd.DataFrame(archivo_datos.json())
```

### Comprobacion incoherencias de datos

```python
##Primero debemos asegurarnos que el customer id sea unico en todos los datos
resultado=len(datos)-len(datos["customerID"].unique())
print(resultado)
## Como el resultado es 0 todos los clientes son distintos
```

```python
## Normalizamos los datos para verfificar los demas datos
informacion_cliente=pd.json_normalize(datos["customer"])
informacion_cuenta=pd.json_normalize(datos["account"])
informacion_numero=pd.json_normalize(datos["phone"])
informacion_internet=pd.json_normalize(datos["internet"])
datos_normalizados=datos.drop(columns=["customer","account","phone","internet"]).join([informacion_cliente,informacion_numero,informacion_internet,informacion_cuenta])
```

```python
datos_normalizados.head()
```

```python
datos_normalizados.info()
```

```python
## Segun el info nos seguramos que no hay datos nulos ni que arreglar o revisar
```

### Manejo de inconsistencias

```python
##nos aseguraremos de que las columnas no tengan valores extraños
```

```python
datos_a_verificar=datos_normalizados.drop(columns=["customerID","Charges.Monthly","Charges.Total"])
```

```python
for columna in datos_a_verificar.columns.tolist():
    print(columna)
    for valor in datos_a_verificar[columna].drop_duplicates():
        print (valor)
```

```python
## Como podemos apreciar la columna churn tiene datos vacios, lo rellenaremos con No
datos_normalizados["Churn"]=datos_normalizados["Churn"].str.capitalize().replace("","No")
```

```python
##A partir de este momento podemos notar algunas inconsistencias, como que hay clientes que tienen 0 meses contratados, los verificaremos
filtro=datos_normalizados["tenure"]==0
verificacion_datos_0_meses=datos_normalizados[filtro].loc[:,["customerID","tenure","Contract","Charges.Monthly","Charges.Total"]]
verificacion_datos_0_meses.head()
```

```python
## Pudimos comprobarlo, los veremos despues como realizarlos, por ahora los datos de No internet service y No phone service los convertiremos en No
```

```python
lista_columnas=["OnlineSecurity",
"OnlineBackup",
"DeviceProtection",
"TechSupport",
"StreamingTV",
"StreamingMovies"
               ]
for columna in lista_columnas:
    datos_normalizados[columna]=datos_normalizados[columna].str.replace("No internet service","No")
datos_normalizados["MultipleLines"]=datos_normalizados["MultipleLines"].str.replace("No phone service","No")
```

### Cuenta diaria

```python
datos_normalizados["Cuenta diaria"]=round((datos_normalizados["Charges.Monthly"]/30),3)
```

```python
datos_normalizados.head()
```

### Estandarizacion de datos

```python
### Como anteriormente hicimos con el No internet service y No phone service ahora convertiremos esos datos en 1 y 0, primero identificaremos las columnas que contienen estos datos
```

```python
datos_normalizados.dtypes
```

```python
columnas_a_modificar = datos_normalizados.columns[datos_normalizados.isin(["Yes"]).any()]
```

```python
for x in columnas_a_modificar:
    datos_normalizados[x]=datos_normalizados[x].str.capitalize().replace("Yes","1")
    datos_normalizados[x]=datos_normalizados[x].str.capitalize().replace("No","0")
    datos_normalizados[x]=(datos_normalizados[x]).astype(int)
```

### Distribucion de evasion-Recuento de evacion en variables categoricas

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

## Podemos ver que la evasion se da muchisimo mas en los contratos meses a meses.
sns.countplot(x='Contract', hue='Churn', data=datos_a_analizar)
plt.title('Churn por tipo de contrato')
plt.show()
```

```python
## Podemos ver que no hay diferencia entre hombres y mujeres, no es una variable importante
sns.countplot(x='gender', hue='Churn', data=datos_a_analizar)
plt.title('Churn por genero')
plt.show()
```

```python
plt.figure(figsize=(20, 6))  # Ancho=10, Alto=6 pulgadas
sns.countplot(x='tenure', hue='Churn', data=datos_a_analizar)
plt.title('Churn por genero')
plt.show()
## Podemos ver que los evasores apenas llevan 1 mes cuando empuezan a evadir, porque a medida que avanza el tiempo no se vencantidades de evasores significantes
```

```python
## Podemos ver que la mayoria de evasores lo hacen por cheque electronico
plt.figure(figsize=(20, 6))  # Ancho=10, Alto=6 pulgadas
sns.countplot(x='PaymentMethod', hue='Churn', data=datos_a_analizar)
plt.title('Churn por genero')
plt.show()
```
